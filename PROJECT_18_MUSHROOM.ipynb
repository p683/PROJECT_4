```python
data.drop(['bruises','gill-color','ring-type','veil-color'],axis=1,inplace=True)
```

```python
data.shape
```

```python
# DIVIDING THE DATA INTO INDEPENDENT AND DEPENDENT VARIABLES
x = data.drop(['class'],axis=1)
y = data['class']
# X == INDEPENDENT AND Y == DEPENDENT
```

```python
# SPLITTING THE DATASET THROUGH CROSS_VAL_SCORE
# IMPORTING THE LIBRARY
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
# APPLYING BAGGING AND BOOSTING BOTH TYPE OF MODELS
```

```python
# BAGGING MODEL
# IMPORTING THE LIBRARIES
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
rf = RandomForestClassifier(n_estimators=20,criterion='entropy',max_depth=5)
dtc = DecisionTreeClassifier(criterion='entropy',max_depth=5)
```

```python
score = cross_val_score(rf,x,y,cv=7,scoring='accuracy').mean()*100
```

```python
score
```

```python
y_pred = cross_val_predict(rf,x,y,cv=7)
```

```python
# SCORE METRICS FOR THE METRICS
# IMPORTING THE LIBRARIES
from sklearn.metrics import confusion_matrix,roc_auc_score,accuracy_score
```

```python
matrix = confusion_matrix(y_pred,y)
matrix
```

```python
score = accuracy_score(y_pred,y)*100
score
```

```python
roc_score = roc_auc_score(y_pred,y)
roc_score
```

```python
############ RANDOM FOREST IS PREDICTING VERY GOOD #########
```

```python
# decision tree
score = cross_val_score(dtc,x,y,cv=7,scoring='accuracy').mean()*100
score
```

```python
pred = cross_val_predict(dtc,x,y,cv=7)
```

```python
matrix = confusion_matrix(pred,y)
matrix
```

```python
# CLEARLY IT DEPICTS THAT DECISION TREE HAS PREDICTED LESS FALSE NEGATIVE RATE 
```

```python
roc_score = roc_auc_score(pred,y)
roc_score
```

```python
######### BOOSTING MODELS ##############
# IMPORTING THE LIBRARIES
from sklearn.ensemble import AdaBoostClassifier
```

```python
boost = AdaBoostClassifier(n_estimators=50)
```

```python
score = cross_val_score(boost,x,y,cv=7,scoring='accuracy').mean()*100
score
```

```python
pred = cross_val_predict(boost,x,y,cv=7)
```

```python
matrix = confusion_matrix(pred,y)
matrix
```

```python
# FALSE NEGATIVE RATE ARE LESS THAN BOTH THE MODELS
```

```python
roc_score = roc_auc_score(pred,y)
roc_score
```

```python
# BASED ON ACCURACY SCORE OF THE MODEL AND FALSE NEGATIVE RATE 
# DECISION TREE IS PREDICTING THE BEST
```
